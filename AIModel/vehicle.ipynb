```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Recreate x after label encoding has been applied to the 'data' DataFrame
req_feat_transport = [
    'Vehicle_Type',
    'Fuel_Type',
    'Engine_Size',
    'Age_of_Vehicle',
    'Mileage',
    'Speed',
    'Acceleration',
    'Road_Type',
    'Traffic_Conditions'
]
x = data[req_feat_transport]

model = RandomForestRegressor(
    n_estimators=400,
    max_depth=12,
    min_samples_split=5,
    random_state=42
)
model.fit(x, y)

feat_importances = pd.Series(model.feature_importances_, index=x.columns)
feat_importances.nlargest(10).plot(kind='barh', color='teal')
plt.title("Top Factors Driving Carbon Emissions")
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, x, y, cv=5, scoring='r2')

print("New CV R2 Scores:", scores)
print("Mean R2:", scores.mean())
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import r2_score

model = XGBRegressor(
    n_estimators=1000,
    learning_rate=0.03,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    early_stopping_rounds=50,
    eval_metric="rmse",
    random_state=42,
    enable_categorical=True # Enable categorical feature support
)

# Explicitly convert categorical features to 'category' dtype for XGBoost
categorical_features = ['Vehicle_Type', 'Fuel_Type', 'Road_Type', 'Traffic_Conditions']
for col in categorical_features:
    if col in x_train.columns:
        x_train[col] = x_train[col].astype('category')
    if col in x_test.columns:
        x_test[col] = x_test[col].astype('category')

model.fit(
    x_train, y_train,
    eval_set=[(x_test, y_test)],
    verbose=False
)

y_pred = model.predict(x_test)

print("R2 Score:", r2_score(y_test, y_pred))
```
